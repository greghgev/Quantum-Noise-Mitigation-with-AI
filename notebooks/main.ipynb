```python
import sys
from pathlib import Path

# 1. Truco MLOps para que el notebook encuentre la carpeta 'src'
# Añadimos la raíz del proyecto (TFM-Quantum) al path de Python
root_path = Path.cwd().parent
if str(root_path) not in sys.path:
    sys.path.append(str(root_path))

from src.quantum_gen import CircuitFactory
from src.config import SEED

# 2. Creamos un 'Coupling Map' falso para estresar a la clase
# Simulamos un chip de 5 qubits en línea recta: 0-1-2-3-4
# Esto forzará al transpiler a meter puertas SWAP si intentamos entrelazar el 0 con el 4.
mock_coupling_map = [
    [0, 1], [1, 0], [1, 2], [2, 1], [2, 3], [3, 2], [3, 4], [4, 3]
]

# 3. Instanciamos nuestra fábrica
print("⚙️ Inicializando CircuitFactory...")
factory = CircuitFactory(min_qubits=5, max_qubits=15)
num_qubits = 5

# --- PRUEBA 1: Circuito Aleatorio ---
print("\n--- 🧪 TEST 1: Circuito Aleatorio ---")
random_qc = factory.generate_random_circuit(
    num_qubits=num_qubits, 
    logical_depth=3, 
    seed=SEED, 
    coupling_map=mock_coupling_map
)
print(f"Profundidad física real: {random_qc.depth()}")
display(random_qc.draw(output='mpl', fold=-1, style='iqp'))

# --- PRUEBA 2: QAOA ---
print("\n--- 🧪 TEST 2: QAOA (Estructurado) ---")
qaoa_qc = factory.generate_qaoa(
    num_qubits=num_qubits, 
    seed=SEED, 
    reps=1, 
    coupling_map=mock_coupling_map
)
print(f"Profundidad física real: {qaoa_qc.depth()}")
display(qaoa_qc.draw(output='mpl', fold=-1, style='iqp'))

# --- PRUEBA 3: Grover ---
print("\n--- 🧪 TEST 3: Algoritmo de Grover ---")
grover_qc = factory.generate_grover(
    num_qubits=num_qubits, 
    coupling_map=mock_coupling_map
)
print(f"Profundidad física real: {grover_qc.depth()}")
display(grover_qc.draw(output='mpl', fold=-1, style='iqp'))
```
